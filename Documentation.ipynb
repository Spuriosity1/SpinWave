Typical usage is guided by the classes present in `SpinWelle.py`. The inheritance structure is straightforward:

`Lattice -> PlottingLattice -> SpinWave`

and all three class constructors take the same arguments. The usage and general structure of the code will be outlined in (roughly) the order it was written, which will hopefully be the easiest to parse.

### Inherited from `Lattice`
This is responsible for adding atoms to the unit cell and managing the connections between them. Currently, atoms cannot be removed from the unit cell - instead, a new  instance must be initialised.

**Class Constructor**
`SpinWave(a1, a2, a1m=None, a2m=None)`
 - `a1, a2`  
     The real space translation vectors of the lattice
 - `a1m, a2m`  
     The real space translation vectors of the magnetic unit cell. These default to a1, a2 if either is missing.
 
**Properties**
 - `area`  
     Lattice unit cell area
 - `mag_area`  
     Magnetic unit cell area
 - `reciprocal_vectors`  
     Orthogonal conjugate vectors to a1, a2. Returns `[b1, b2]` 
     such that $\vec{A}_i \cdot \vec{B}_j = 2\pi \delta_{ij}$
 - `mag_reciprocal_vectors`  
     Orthogonal conjugate vectors to a1m, a2m. Returns `[b1m, b2m]`
     such that $\vec{a}_i \cdot \vec{b}_j = 2\pi \delta_{ij}$

**Methods**
 - `set_dist_units( num, string)`
     Allows you to specify a length scale to measure in units of. E.g. a square lattice with a lattice constant of 3.1Å, you would call 
     `set_dist_units(3.1, 'Å')` then specify the atom positions as (0,0), (1,0), (0,1), (1,1). Note that this number only gets used when plotting - you can change it at will and everything *should* still work. (Try not to do that though ,because I haven't tested it!)
 - `add_atom(label, xy, R=np.eye(3))`  
     Creates an atom with realspace coordinates given by the 2-tuple `xy`, 
     such that the classical ground state is given by `R.(0,0,1)`.
 - `set_rotor(label, R)`  
     Sets the onsite rotation of site `label` to `R`.
 - `define_coupling(name, matrix = np.zeros(3,3))`
     Defines a new kind of coupling, corresponding to a matrix $M$ such that the spin spin interactions it's applied to look like $\mathbf{S}^T_i M \mathbf{S}_j$. Calling this funciton on an for an existing `name` overwrites its associated matrix.
 - `add_coupling(self, label1, label2, coupling_name, delta=(0.01,0.009),color='k')`  
     Adds a bond between atoms corresponding to `label1` and `label2`, with coupling matrix named by `coupling_name`.
     _Subtlety_: the site `label1` is taken to "belong to" the first unit cell, 
     at the position `xy` specified when `add_atom` was called. Due to periodicity, 
     there are infinitely many sites corresponding to `label2`. 
     `delta` specifies a vector from site 1 to site 2 - the link is made to the '2' site closest to `xy + delta`, which might be in another unit cell.
 - `clear_couplings()`  
     Deletes all bonds. (Note that all previously defined couplings retain their previuous values, they just aren't applied to anyhting any more)
 - `set_CGS(label, G)`  
     Sets the rotor at site `label` to a matrix that rotates `(0,0,1)` to `G`. Note that this matrix is not unique, but ultimately it shakes out as an irrelevant phase.
 - `get_CGS(label)`  
     Returns the vector `R.(0,0,1)` for the specified site
     

### Inherited from `PlottingLattice`
This layer is responsible for plotting real-space and k-space representations of the lattice.

**Properties**
 - `fig_RS`  
     Figure in which real-space representations are displayed
 - `ax_RS`  
     matplotlib `Axes3D` object for real-space plotting 
 - `fig_RS`  
     Figure in which $k$-space representations are displayed
 - `ax_RS`  
     matplotlib `Axes` object for 2D $k$-space plotting


**Methods**
 - `plot_atoms(self,n1=1,n2=1, arrowlen=0.5, plot_GS=True, plot_labels=True)`  
    Plots all the atoms added to the unit cell in real space, tiled out in a `n1` by `n2` grid.  
    `plot_GS` flags whether or not to plot classical ground state vectors, which are scaled by `arrowlen`.
    `plot_labels` flags whether or not to label sites.
 - `plot_couplings(self, n1=1,n2=1, bond_width=1,style='decompose',scale=1)`  
     Plots the connections between atoms, tiled out in a `n1` by `n2` grid.  
     `bond_width` changes the length of lines connecting atoms.  
     `style` may be either `net` or `eigs`. `net` plots bonds only, `eigs` 
     plots an ellipsoid representative of coupling strength in spin-space. 
     (The semi-major axes are oriented along the eigenvectors of the $J$ tensor, of lengths given by the eigenvalues)
     `scale` scales this ellipsoid.
 - `plot_BZ(self, unitcell="primitive", plot_MBZ=True, plot_LBZ=True, *args, **kwargs)`  
     Plots the Brillouin zone specified in the `unitcell` argument. If it is a string, it will plot a polygon with vertices
     - primitive: `[(1,1), (1, -1), (-1, -1), (-1, 1)]/2`
     - hexagon: `[(1,1), (-1, 2), (-2, 1), (-1, -1), (1, -2), (2, -1)]/3`
     - hexagon2: `[(2,1), (1, -1), (-1, -2), (-2, -1), (-1, 1), (1, 2)]/3`  
     where the vectors are in **reciprocal lattice units**. If `unitcell` is a list or numpy array, the program attempts to use it directly as the list of vertices.
    `plot_MBZ` and `plot_LBZ` flag the plotting of the magnetic and lattice Brillouin zones. Note that the MBZ plotting is hardcoded to plot the primitive magnetic zone.  
    `*args, **kwargs` are passed to `ax_KS.plot`.   
 - `plot_BZ_path(self, path, *args, **kwargs)`  
     Plots a path specified as a dictionary of points indexed by labels, e.g.
    `{"X": [1/2, -1/2],"Γ ": [0,0],"Y": [1/2, 1/2]}`
     where the coordinates are again in reciprocal lattice units.  
     `*args, **kwargs` are passed to `ax_KS.plot`.
 
 
     
     

### Inherited from `SpinWave`

**Properties**
- `spinspace_basis` A matrix containing the three orthogonal basis vectors for the spin axes, which are stored as **rows**. Defaults to the standard basis.
- `quiet` A boolean flag that suppresses some error messages.
- `spinham` A $3n$ by $3n$ matrix $M$ such that the classical energy of the system is given by $E_{classical} = (S_1^T, S_2^T, ... S_n^T) . M . \begin{pmatrix}S_1\\S_2\\ \vdots \\ S_n \end{pmatrix}$ 

**Methods**
- `set_spinspace_basis(O)` Setter for `spinspace_basis`. Accepts a matrix `O = [v1, v2, v3]`. Checks that the vectors you provide are orhtogonal, normalised and right handed. **note that this transposes the matrix you provide!**
- `get_Zmat(alpha, beta)` Returns the tensor corresponding to $Z^{\alpha\beta}_{ij}$, as defined in the notes.
- `get_z(label)` Returns the complex $z^\alpha$ vector $R^{\alpha 1} + i R^{\alpha 2}$ for the site with label `label`.
- `calc_Hamiltonaian(K)` Accepts a vector of **2D** $k$ space coordinates with shape $(M, 2)$, returning a sequence of numerical Bloch Hamiltonians with shape $M, 2n, 2n$ where $n$ is the number of operators in the unit cell. 
- `first_order_terms()` Returns the $n$ coefficients of the first order terms as a vector.
- `energy(K, tol=1e-9)` Calculates the energy eigenvalues of the bosonic Bloch Hamiltonian along the $K$ space path you specify, in the shape `(M, 2)`. Returns a matrix with dimensions `(M, 2n)`, where $n$ is the number of operators in the unit cell. This method is faster than `Colpa_diagonalise` and `Bogo_diagonalise`, but it does not tell you the matrices responsible for the Bogoliubov transform.
- `isgapped(K), mincond=100` Returns a boolean array of length `M` based on a $k$ space path as for calc_Hamiltonian, corresponding to whether or not the condition number of the eigenvalue problem is larger than mincond. This measures whether the matrix is singular, i.e. `True` entries correspond to nonsingular matrices.
- `isposdef(K)` Checks if the Hamiltonian is positive definite along the walk you provide.
- `Colpa_diagonalise(K)` Implements the Colpa algorithm as written in Appendix A of Smit et al., 'Magnon daming in the zigzag phase of the Kitaev-Heisenberg-Γ model on a honeycomb lattice' 2020.
    Returns the `M, 2n` energy vector $E$ in ascending order, and the corresponding Bogoliubov transforms on the extended space $B$ of size `M, 2n, 2n`.
- `Bogo_diagonalise(K)` dpes the same thing as `Colpa_diagonalise(K)` using a different numerical method that is less sensitive to positive definiteness, but also less numerically stable.
- `get_ffactor(K, alpha, beta)` returns a vector of energies $E$ and spin correlation tensor elements $\langle S^\alpha S^\beta \rangle$, of sizes of `M, 2n` and  `M, 2n, 2n` respectively. See notes for details.
- `get_perp_ffactor(self, K)` Accepts a list of 2- or 3-vectors for `K`. Returns the enegies $E$ and the tensor sum $\sum_{\alpha \beta} \left(\delta_{\alpha\beta} - \frac{K_\alpha K_\beta}{|\mathbf{K}|^2}\right)\langle S^\alpha S^\beta \rangle$. Note that I assume a flat dispersion in the z axis. These have the shapes `M, 2n` and `M, 2n`. 


# The PowderAverage class

## Constructor

`PowderAverage(spinw, gauss_broaden, intrinsic_broaden)`
This should be renamed to something else, more like 'SpectralWeight'. Icbb now.
The Dirac delta like excitation spectra need to be empirically broadened to account for the finite detector resolution and the finite lifetime of excited states. This is managed by convolution with a Voigt profile, which is itselt the covolution of a Gaussian with a Lorentzian. The broadening parameters for these elementary distributions are `gauss_broaden` and `intrinsic_broaden` respectively.

`spinw` should be a SpinWave object storing all of the relevant parameters. Note that only a shallow copy is stored - changes to the spinw object are sill 'seen' by this class.

## Methods
- `set_form_factor(func)` Sets the (isotropic) form factor corresponding to the spatial extent of the orbital to the function `func(Q)`.
- `set_form_factor_arr(Q, f)` Accepts two arrays of equal length `Q` and `f`, and sets the internal form factor to a cubic interpolation of the function you specify.
- `plot_single_crystal(path, Egrid, ab_index, use_ffactor=False, plot_dispersion=True, lw=0.3)`
Plots the scattering strength along a path, in the same format as `plot_BZ_path`. `ab_index` can be any of `xx, xy, xz, yx, yy, yz, zx, zy, zz` or `perp`, and it indicates which correlations to plot. `plot_dispersion` overlays a thin dashed white line (of width `lw`) over the colour plot. `Egrid` is an array of energies over which to evaluate the spectral weight, I recommend `np.linspace(0,e_max, 100)` where `e_max` is the highest energy you want to look at.
- `calc_paverage(Egrid, Qgrid, num_samples=500)` Performs a Monte Carlo  integration over the spherical shells `(Qgrid[n], Qgrid[n+1])`. `Egrid` is again the grid of energies you want to look at. `num_samples` is the number of points to integrate over per shell.
- `plot_paverage(logscale=False, vmin=None, vmax=None, cmap='jet', use_ffactor=False, ax=None, **kwargs)`
These handles do what you think they do.`vmin, vmax, cmap` are passed directly to `ax.pcolormesh`, `ax` allows you to specify your own Axes (otherwise the function makes its own), and `**kwargs` don't do anything. Note that this only works if you've already calculated the simulated scattering data by calling `calc_paverage`.
- `plot_cut(self, Q, Egrid, npoints=1000, dQ=0, use_ffactor=True, **kwargs)`
Does not need `calc_paverage` to be run. This calculates a single Q cut (of width `dQ`) of the powder averaged scatterign data.


```python
%matplotlib inline
import numpy as np
from numpy import linalg
import matplotlib.pyplot as plt

from SpinWelle import SpinWave, PowderAverage
from rotation import Rx, Ry, Rz
```

The idea here is to create a "shell" of a model, then decide on values of parameters and the classical ground state later.

```python
# Zig-zag state on a honeycomb lattice
# Bare minimum scaffolding

# Create a SpinWave object:
lat = SpinWave([np.sqrt(3)/2,1.5],[-np.sqrt(3)/2,1.5], [np.sqrt(3), 0], [0,3])
lat.set_dist_units(3.0957,'Å')

# make 4 unique sites. These labels are your your benefit, they can be any string
lat.add_atom('a',(0,1))
lat.add_atom('b',(-np.sqrt(3)/2,0.5))
lat.add_atom('c',(-np.sqrt(3)/2,-0.5))
lat.add_atom('d',(0,-1))

lat.plot_atoms()
# This shows the  default z-facing ground state
```

```python
# An important technicality: The set_spin_basis function
# you might want to use different axes to specify the spin-spin interactions. (This is the convention used for Kitaev models!)
e1 = np.array([np.sqrt(1/6), -np.sqrt(1/2), np.sqrt(1/3)])
e2 = np.array([np.sqrt(1/6), np.sqrt(1/2), np.sqrt(1/3)])
e3 = np.array([-np.sqrt(2/3), 0, np.sqrt(1/3)])

lat.set_spin_basis([e1,e2,e3])
lat.plot_atoms()
# This shows the  default z-facing ground state, but now "z" is the Kitaev "z" direciton
```

```python
# Coupling definition. Can also provide an initial matrix value, e.g.
# lat.define_coupling('Jx', np.array([[1,1,0],[1,1,0],[0,0,-3]])
# As written, defaults to zero (this will be updated later)
# The labels can be whatever string you want, they're just placeholders.
lat.define_coupling('Jx')
lat.define_coupling('Jy')
lat.define_coupling('Jz')
lat.define_coupling('J2')
lat.define_coupling('J3')


lat.clear_couplings()

# Define the nearest-neighbour (matrix) couplings
lat.add_coupling('a', 'b', 'Jx', (-1, 0),color='green')
lat.add_coupling('d', 'c', 'Jx', (1, 0),color='green')

lat.add_coupling('a', 'b', 'Jy', (1, 0),color='red')
lat.add_coupling('c', 'd', 'Jy', (1, 0),color='red')

lat.add_coupling('b', 'c', 'Jz',(0,-1), color='blue')
lat.add_coupling('a', 'd', 'Jz',(0, 1), color='blue')

# Define the third-nearest-neighbour Heisenberg couplings
lat.add_coupling('a','d', 'J3', (0,-2), color='k')
lat.add_coupling('b','c', 'J3', (2,-1), color='k')
lat.add_coupling('c','b', 'J3', (2,1), color='k')

lat.add_coupling('b','c', 'J3', (0,2), color='k')
lat.add_coupling('a','d', 'J3', (-2,1), color='k')
lat.add_coupling('a','d', 'J3', (2,1), color='k')

r32 = np.sqrt(3)/2

# Define the second-nearest-neighbour Heisenberg couplings
lat.add_coupling('a','a', 'J2', (2*r32,0), color='orange')
lat.add_coupling('b','b', 'J2', (2*r32,0), color='orange')
lat.add_coupling('c','c', 'J2', (2*r32,0), color='orange')
lat.add_coupling('d','d', 'J2', (2*r32,0), color='orange')

lat.add_coupling('a','c', 'J2', (r32,1.5), color='orange')
lat.add_coupling('a','c', 'J2', (-r32,-1.5), color='orange')
lat.add_coupling('b','d', 'J2', (r32,1.5), color='orange')
lat.add_coupling('d','b', 'J2', (r32,1.5), color='orange')

lat.add_coupling('a','c', 'J2', (r32,-1.5), color='orange')
lat.add_coupling('a','c', 'J2', (-r32,1.5), color='orange')
lat.add_coupling('b','d', 'J2', (r32,-1.5), color='orange')
lat.add_coupling('d','b', 'J2', (r32,-1.5), color='orange')


# Show them!
lat.plot_couplings(3,3)
```

```python
# Helper functions
def Jx(J, K, G, Gp, **args):
    return np.array([[J+K, Gp, Gp],
                     [Gp,  J,  G ],
                     [Gp,  G,  J]])
def Jy(J, K, G, Gp, **args):
    return np.array([[J,   Gp,  G ],
                     [Gp, J+K, Gp ],
                     [G,   Gp,  J ]])
def Jz(J, K, G, Gp, **args):
    return np.array([[ J,   G,  Gp  ],
                     [ G,   J,  Gp  ],
                     [Gp,  Gp,  J+K ]])

# I recommend modifying this funciton for your purposes.
# This takes in a dictionary with keys K,J,G,Gp,J2,J3 and modifies / redefines all of the couplings to match.
def build_model(params):
    
    # HACK: This magic matrix just happens to always have its smallest eigenvalue point in the CGS direction. This is not true in general!
    ene, v = np.linalg.eigh( Jx(**params) + Jy(**params) - Jz(**params) - 3*params['J3']*np.eye(3) - 2*params['J2']*np.eye(3))
    v = v[:, np.argmin(ene)]
    
    lat.set_CGS('a', v)
    lat.set_CGS('b', v)
    lat.set_CGS('c', -v)
    lat.set_CGS('d', -v)
    
    # Update all of the couplings
    lat.define_coupling('Jx', 0.5*Jx(**params))
    lat.define_coupling('Jy', 0.5*Jy(**params))
    lat.define_coupling('Jz', 0.5*Jz(**params))
    lat.define_coupling('J2', 0.5*params['J2']*np.eye(3))
    lat.define_coupling('J3', 0.5*params['J3']*np.eye(3))
```

```python
special_points_zz = {
    "X": [-1/2, 1/2],
    "Γ ": [0,0],
    "Y": [1/2, 1/2],
    "Γ'": [1,0],
    "M" : [1/2,0],
    "Γ  ": [0, 0]
}
lat.plot_BZ('hexagon2')
lat.plot_BZ_path(special_points_zz)
for side in lat.ax_KS.spines.values():
    side.set_visible(False)
lat.ax_KS.axis('off')
lat.ax_KS.spines['bottom'].set_visible(False)
lat.ax_KS.spines['bottom'].set_visible(False)

lat.plot_bandgraph(special_points_zz)
```

```python
# Doing spherical averages

spec = PowderAverage(lat)
# set the broadening parameters (Voigt profile)
spec.intrinsic_broaden = 0.2
spec.gauss_broaden = 0.135

# Set up the grid to evaluate eveything on
Egrid = np.linspace(0,12,100)
Qgrid = np.linspace(0.01,4.65,100)

# Monte Carlo integrate the powder average (time consuming for alrge num_samples)
spec.calc_paverage(Egrid, Qgrid, num_samples=500)
spec.plot_paverage(logscale=False,vmax=2.2,vmin=0, use_ffactor=False)
```

```python
# You can also plot 'cuts' of single Q values:
spec.ax_cuts=None
spec.plot_cut(0.3,Egrid,use_ffactor=False)
spec.plot_cut(0.75,Egrid,use_ffactor=False)
spec.plot_cut(2.75,Egrid,use_ffactor=False)
```
